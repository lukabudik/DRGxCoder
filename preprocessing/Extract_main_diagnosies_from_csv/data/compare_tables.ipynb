```python
import pandas as pd
import numpy as np
```

```python
#Load the first table
data_main = pd.read_csv('drg_diagnozy_main_extracted_filtered.csv')
data_main_vs_secondary = pd.read_csv('drg_diagnozy_main_vs_secondary_extracted_filtered.csv')
```

```python
print(f"Main has {data_main.shape[0]} rows and {data_main.shape[1]} columns: {data_main.columns.tolist()}")
print(f"Main vs Secondary has {data_main_vs_secondary.shape[0]} rows and {data_main_vs_secondary.shape[1]} columns: {data_main_vs_secondary.columns.tolist()}")
```

```python
# Compare overlaps of Kod and Kod VDG
main_kods = set(data_main['Kod'])
main_vs_secondary_kods = set(data_main_vs_secondary['Kod VDG'])
overlap_kods = main_kods.intersection(main_vs_secondary_kods)
print(f"Number of overlapping Kod VDG: {len(overlap_kods)}")
# See unique values in Kod VDG not in Kod
unique_to_secondary = main_vs_secondary_kods - main_kods
print(f"Number of Kod VDG unique to main vs secondary: {len(unique_to_secondary)}")
# See unique values in Kod not in Kod VDG
unique_to_main = main_kods - main_vs_secondary_kods
print(f"Number of Kod unique to main: {len(unique_to_main)}")
print("Sample Kod VDG unique to main vs secondary:", list(unique_to_secondary)[:10])
print("Sample Kod unique to main:", list(unique_to_main)[:10])
# Duplicates in Kod VDG
duplicates_in_secondary = data_main_vs_secondary['Kod VDG'][data_main_vs_secondary['Kod VDG'].duplicated()]
print(f"Number of duplicate Kod VDG in main vs secondary: {duplicates_in_secondary.nunique()}")
print("Sample duplicate Kod VDG in main vs secondary:", duplicates_in_secondary.unique()[:10])

# Inersection of Kod HDG with Kod from main
kod_hdg_set = set(data_main_vs_secondary['Kod HDG'])
intersection_hdg_main = kod_hdg_set.intersection(main_kods)
print(f"Number of Kod HDG that are also in main Kod: {len(intersection_hdg_main)}")
# Kods missing in Kod HDG
missing_in_hdg = main_kods - kod_hdg_set
print(f"Number of Kod in main missing in Kod HDG: {len(missing_in_hdg)}")
print("Sample Kod in main missing in Kod HDG:", list(missing_in_hdg)[:10])
```
